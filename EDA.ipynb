```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("training_data/Full_review_QnA_intent_processed_amazon_reviews_us_Beauty.tsv", sep="\t", quoting=csv.QUOTE_NONE)
```

```python
# Set up the aesthetics for the plots
sns.set(style="whitegrid")
```

```python
# 1. Distribution of Star Ratings
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='star_rating', palette='viridis')
plt.title('Distribution of Star Ratings - Health & Personal Care Products')
plt.xlabel('Star Rating')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Number of Reviews per Product (Top 10)
top_products = df['product_title'].value_counts().head(10)
plt.figure(figsize=(12, 8))
top_products.plot(kind='bar', color='skyblue')
plt.title('Top 10 Health & Personal Care Products by Number of Reviews')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Product Title')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
# 3. Helpful Votes Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='helpful_votes', bins=30, color='salmon', kde=True)
plt.title('Distribution of Helpful Votes - Health & Personal Care Products')
plt.xlabel('Number of Helpful Votes')
plt.ylabel('Count')
plt.xlim(0, df['helpful_votes'].quantile(0.95))  # Limiting to 95th percentile for better visibility
plt.show()
```

```python
# 4. Verified Purchase Reviews
verified_counts = df['verified_purchase'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(verified_counts, labels=verified_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'lightcoral'])
plt.title('Verified Purchase Reviews Distribution - Health & Personal Care Products')
plt.show()
```

```python
# 5. Vine Reviews Distribution
vine_counts = df['vine'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(vine_counts, labels=vine_counts.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'purple'])
plt.title('Vine Reviews Distribution - Health & Personal Care Products')
plt.show()

```

```python
# 6. Star Ratings vs. Helpful Votes
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='star_rating', y='helpful_votes', alpha=0.6, edgecolor=None, color='blue')
plt.title('Star Ratings vs. Helpful Votes - Beauty Products')
plt.xlabel('Star Rating')
plt.ylabel('Helpful Votes')
plt.ylim(0, df['helpful_votes'].quantile(0.95))  # Limiting to 95th percentile for better visibility
plt.show()
```
